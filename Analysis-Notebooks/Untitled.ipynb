```python
def fizzBuzzy(x, y):
    for i in range(1, 100):
        if i % x == 0:
            print("Fizz:" + str(i))
        if i % y == 0:
            print("Buzz: " + str(i))
        if i % 3 == 0 and i % 5 == 0:
            print("FizzBuzz: " + str(i))
        else: 
            print(i)
        
fizzBuzzy(3, 5)
```
